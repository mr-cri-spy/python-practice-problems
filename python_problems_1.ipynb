Write a program that takes as input.
Using conditional statements,calculate the based on these rules : Q1 salary final tax rate • If salary<30,000 → 5% • If salary is 30,000 – 70,000 → 15% • If salary > 70,000→25%

```python
salary = float(input("Enter your salary: "))
```

```python
if salary < 30000:
  tax_rate = 0.05
elif salary >=30000 and salary <= 70000:
  tax_rate = 0.15
else :
  tax_rate = 0.25

tax = salary*tax_rate*100/100
```

```python
print("the salary is : ", salary)
print("tax_rate", tax_rate*100, "%")
print("tax_amount", tax)
```

Write a function that takes two integers and and print sall even numbers between them (inclusive)

```python
number1 = int(input("enter the number"))
number2 = int(input("enter the number"))
```

```python
def print_even_numbers(number1,number2):
  for num in range(number1, number2+1):
    if num % 2 ==0:
      print(f" {num}")

```

```python
print_even_numbers(number1, number2)
```

Write a function to return the the number of digits in a number count

```python
n = int(input("enter the number : "))
def count_digit(n):
  count =0
  while n>0:
    n = n//10
    count += 1
  return count
```

```python
print(count_digit(125437))
```

Write a function to return the of a number sum of digits of n

```python
def sum_of_digits(n):
  sum =0
  while n>0:
    digit = n%10
    sum+=digit
    n = n // 10
  return sum

sum_of_digits(3490)
```
