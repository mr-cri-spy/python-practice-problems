Convert raw user input → clean → extract info → send to LLM → return smart response

input cleaning engine

```python
import re

def clean_text(text):
    text = text.lower()

    # slang normalization
    text = re.sub(r'\bu\b', 'you', text)
    text = re.sub(r'\br\b', 'are', text)

    # remove noise
    text = re.sub(r'[^a-z0-9\s@.+-]', '', text)
    text = re.sub(r'\s+', ' ', text)

    return text.strip()
```

Entity extraction

```python
def extract_entities(text):
    entities = {}

    # email
    email = re.findall(r'[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-z]{2,}', text)
    if email:
        entities['email'] = email

    # phone
    phone = re.findall(r'(?:\+91[-\s]?)?[6-9]\d{9}', text)
    if phone:
        entities['phone'] = phone

    # numbers (orders, ids)
    numbers = re.findall(r'\d+', text)
    if numbers:
        entities['numbers'] = numbers

    return entities
```

```python
#intent detection simple rule based


def detect_intent(text):
    if "order" in text:
        return "order_query"
    elif "price" in text:
        return "pricing"
    elif "book" in text:
        return "booking"
    else:
        return "general"
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

def generate_response(cleaned_text, intent, entities):
    prompt = f"""
    Intent: {intent}
    Entities: {entities}
    User: {cleaned_text}
    Response:
    """

    result = generator(prompt, max_length=100)
    return result[0]['generated_text']
```

import requests

def generate_response(cleaned_text, intent, entities):
    prompt = f"Intent: {intent}, Entities: {entities}, User: {cleaned_text}"

    response = requests.post(
        "https://openrouter.ai/api/v1/chat/completions",
        headers={
            "Authorization": "Bearer YOUR_API_KEY"
        },
        json={
            "model": "mistralai/mixtral-8x7b",
            "messages": [{"role": "user", "content": prompt}]
        }
    )

    return response.json()['choices'][0]['message']['content']

```python
#full pipeline

def chatbot_pipeline(user_input):
    cleaned = clean_text(user_input)
    entities = extract_entities(cleaned)
    intent = detect_intent(cleaned)
    response = generate_response(cleaned, intent, entities)

    return response


# Test
user = "Hey!!! I want to book a table, my number is 9876543210"
print(chatbot_pipeline(user))
```
